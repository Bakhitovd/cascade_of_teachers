```python
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS
from chatterbox.mtl_tts import ChatterboxMultilingualTTS
```

```python

audio_prompt_path = "temp/audio_samples/speach_sample_1_min_en_cut_1.0-15.0.wav" 
multilingual_model = ChatterboxMultilingualTTS.from_pretrained(device="cuda")
text_to_speak = "Съешь еще этих мягких французских булок, да выпей чаю."
wav = multilingual_model.generate(text_to_speak, language_id="RU",
                     audio_prompt_path=audio_prompt_path,
                     exaggeration=0.9)  # ← Drama boost!
ta.save("test-tts_10sec.wav", wav, multilingual_model.sr)
```
