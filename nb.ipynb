```python
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS
from chatterbox.mtl_tts import ChatterboxMultilingualTTS
```

```python

audio_prompt_path = "temp/audio_samples/speach_sample_1_min_en_cut_1.0-15.0.wav" 
multilingual_model = ChatterboxMultilingualTTS.from_pretrained(device="cuda")
text_to_speak = "Съешь еще этих мягких французских булок, да выпей чаю."
wav = multilingual_model.generate(text_to_speak, language_id="RU",
                     audio_prompt_path=audio_prompt_path,
                     exaggeration=0.9)  # ← Drama boost!
ta.save("test-tts_10sec.wav", wav, multilingual_model.sr)
```

```python
# Load model (GPU recommended for speed)
# Ensure you have PyTorch installed with CUDA support if using "cuda"
# You can also use "cpu" but it will be slower.
model = ChatterboxTTS.from_pretrained(device="cuda" if torch.cuda.is_available() else "cpu")
# Text to synthesize
text_to_speak = "Hello, this is a test of Chatterbox voice cloning."
# Path to your 5-second (or more) audio sample
audio_prompt_path = "temp/audio_samples/speach_sample_1_min_ru.wav" # Replace with your .wav file
# Generate speech with voice cloning and emotion
# The 'exaggeration' parameter controls the emotional intensity.
# Values typically range from 0.3 (more monotone) to 0.7 (more dramatic).
wav = model.generate(text_to_speak,
                     audio_prompt_path=audio_prompt_path,
                     exaggeration=0.6)  # ← Drama boost!
# Save the generated audio
output_filename = "cloned_voice_output.wav"
ta.save(output_filename, wav.squeeze(0).cpu(), model.sr)
print(f"Generated audio saved to {output_filename}")
```
